```python
import tensorflow as tf
import tensorflow_hub as hub
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# Define helper function to draw bounding boxes
def draw_boxes(image, boxes, class_names, scores, threshold=0.5):
    """
    Draw bounding boxes and class names on the image.
    """
    image = image.copy()
    h, w, _ = image.shape

    for box, class_name, score in zip(boxes, class_names, scores):
        if score < threshold:
            continue
        ymin, xmin, ymax, xmax = box
        (left, right, top, bottom) = (int(xmin * w), int(xmax * w), int(ymin * h), int(ymax * h))
        
        # Draw rectangle
        cv2.rectangle(image, (left, top), (right, bottom), (0, 255, 0), 2)

        # Add label
        label = f"{class_name.decode('utf-8')}: {score:.2f}"
        cv2.putText(image, label, (left, top - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
    
    return image
```

# Load the Faster R-CNN model from TensorFlow Hub


```python
# Load Faster R-CNN Model
module_handle = "https://tfhub.dev/google/faster_rcnn/openimages_v4/inception_resnet_v2/1"
print("Loading model...")
detector = hub.load(module_handle).signatures['default']
print("Model loaded successfully!")
```

```python
# Load and preprocess image
def load_image(image_path):
    """
    Loads an image and preprocesses it for inference.
    """
    img = cv2.imread(image_path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img_tensor = tf.image.convert_image_dtype(img_rgb, tf.float32)[tf.newaxis, ...]
    return img_rgb, img_tensor

image_path = "education.jpg"
image_rgb, input_tensor = load_image(image_path)
```

```python
# Run inference
print("Running inference...")
prediction = detector(input_tensor)

# Extract results
num_objects = len(prediction["detection_scores"])
scores = prediction["detection_scores"].numpy()
classes = prediction["detection_class_entities"].numpy()
boxes = prediction["detection_boxes"].numpy()

print(f"Found {num_objects} objects.")

# Visualize results
image_with_boxes = draw_boxes(image_rgb, boxes, classes, scores)
plt.figure(figsize=(12, 12))
plt.imshow(image_with_boxes)
plt.axis("off")
plt.show()
```

# Single-Shot Detectors (SSD)

```python
# COCO Label Map for SSD
COCO_LABELS = {
    1: "person", 2: "bicycle", 3: "car", 4: "motorcycle", 5: "airplane",
    6: "bus", 7: "train", 8: "truck", 9: "boat", 10: "traffic light",
    11: "fire hydrant", 13: "stop sign", 14: "parking meter", 15: "bench",
    16: "bird", 17: "cat", 18: "dog", 19: "horse", 20: "sheep",
    21: "cow", 22: "elephant", 23: "bear", 24: "zebra", 25: "giraffe",
    27: "backpack", 28: "umbrella", 31: "handbag", 32: "tie", 33: "suitcase",
    34: "frisbee", 35: "skis", 36: "snowboard", 37: "sports ball", 38: "kite",
    39: "baseball bat", 40: "baseball glove", 41: "skateboard", 42: "surfboard",
    43: "tennis racket", 44: "bottle", 46: "wine glass", 47: "cup", 48: "fork",
    49: "knife", 50: "spoon", 51: "bowl", 52: "banana", 53: "apple",
    54: "sandwich", 55: "orange", 56: "broccoli", 57: "carrot", 58: "hot dog",
    59: "pizza", 60: "donut", 61: "cake", 62: "chair", 63: "couch",
    64: "potted plant", 65: "bed", 67: "dining table", 70: "toilet",
    72: "TV", 73: "laptop", 74: "mouse", 75: "remote", 76: "keyboard",
    77: "cell phone", 78: "microwave", 79: "oven", 80: "toaster", 81: "sink",
    82: "refrigerator", 84: "book", 85: "clock", 86: "vase", 87: "scissors",
    88: "teddy bear", 89: "hair drier", 90: "toothbrush"
}

# Helper function to draw bounding boxes
def draw_boxes(image, boxes, class_ids, scores, threshold=0.5):
    """
    Draw bounding boxes with class labels and scores.
    """
    image = image.copy()
    h, w, _ = image.shape

    for box, class_id, score in zip(boxes, class_ids, scores):
        if score < threshold:
            continue
        ymin, xmin, ymax, xmax = box
        left, right, top, bottom = int(xmin * w), int(xmax * w), int(ymin * h), int(ymax * h)

        # Get class name from COCO labels
        class_name = COCO_LABELS.get(int(class_id), "Unknown")

        # Draw bounding box
        cv2.rectangle(image, (left, top), (right, bottom), (0, 255, 0), 2)

        # Add label
        label = f"{class_name}: {score:.2f}"
        cv2.putText(image, label, (left, top - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    return image
```

```python
# Load SSD MobileNet V2 model from TensorFlow Hub
ssd_model_url = "https://tfhub.dev/tensorflow/ssd_mobilenet_v2/2"
print("Loading SSD model...")
ssd_detector = hub.load(ssd_model_url)
print("SSD model loaded successfully!")
```

```python
# Load and preprocess input image
def load_image(image_path):
    img = cv2.imread(image_path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img_tensor = tf.convert_to_tensor(img_rgb, dtype=tf.uint8)[tf.newaxis, ...]
    return img_rgb, img_tensor

# Run SSD inference
def run_inference(image_path):
    image_rgb, input_tensor = load_image(image_path)
    print("Running SSD inference...")
    results = ssd_detector(input_tensor)

    # Extract results
    boxes = results['detection_boxes'][0].numpy()
    scores = results['detection_scores'][0].numpy()
    class_ids = results['detection_classes'][0].numpy()

    # Draw results
    image_with_boxes = draw_boxes(image_rgb, boxes, class_ids, scores)
    plt.figure(figsize=(12, 12))
    plt.imshow(image_with_boxes)
    plt.axis("off")
    plt.show()

# Path to your image
image_path = "iman.jpg" 
run_inference(image_path)
```

# YOLO (You Only Look Once)

YOLO is a family of real-time object detection models that process an image in a single forward pass of a neural network. Unlike traditional object detection methods, which involve region proposals or sliding windows, YOLO divides the image into a grid and predicts bounding boxes and class probabilities simultaneously.

```python
pip install ultralytics opencv-python matplotlib
```

```python
from ultralytics import YOLO
```

```python
# Load the YOLOv8 pre-trained model
model = YOLO('yolov8n.pt')  # 'yolov8n.pt' is the smallest and fastest model
```

```python
# Load an image
image_path = 'hero.jpg'  # Replace with your image path
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Run inference
results = model(image)

# Extract and visualize results
def visualize_yolo_results(results, image):
    """
    Visualize the YOLO predictions on the input image.
    """
    image_copy = image.copy()
    h, w, _ = image.shape

    # Iterate through detections
    for result in results:
        boxes = result.boxes.xyxy.cpu().numpy()  # Bounding box coordinates
        class_ids = result.boxes.cls.cpu().numpy()  # Class IDs
        confidences = result.boxes.conf.cpu().numpy()  # Confidence scores

        # Draw bounding boxes and labels
        for box, cls_id, conf in zip(boxes, class_ids, confidences):
            x1, y1, x2, y2 = map(int, box)
            label = f"{model.names[int(cls_id)]}: {conf:.2f}"
            cv2.rectangle(image_copy, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(image_copy, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    return image_copy

# Visualize results
image_with_boxes = visualize_yolo_results(results, image_rgb)
plt.figure(figsize=(10, 10))
plt.imshow(image_with_boxes)
plt.axis("off")
plt.show()
```

 # Real-Time Detection!

```python
import cv2
from ultralytics import YOLO

# Load the YOLOv8 model
model = YOLO("yolov8n.pt")  # YOLOv8 Nano model

# Path to your video file
video_path = "video.mp4"  # Replace with your video file path
output_path = "output_yolo_detection.mp4"  # Output video file path

# Open the video file
cap = cv2.VideoCapture(video_path)

# Get video details for saving output
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec for output file
fps = cap.get(cv2.CAP_PROP_FPS)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

# Run real-time detection
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Run inference on the frame
    results = model(frame)

    # Visualize the detections
    annotated_frame = results[0].plot()  # Draw bounding boxes and labels
    out.write(annotated_frame)  # Write to output video

    # Display the frame
    cv2.imshow("YOLOv8 Real-Time Detection", annotated_frame)

    # Press 'q' to quit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
out.release()
cv2.destroyAllWindows()
print("Detection completed and saved to:", output_path)
```

Real-Time Detection Using Webcam

```python
import cv2
from ultralytics import YOLO

# Load the YOLOv8 model
model = YOLO("yolov8n.pt")

# Open the webcam (0 = default camera)
cap = cv2.VideoCapture(0)

# Run real-time detection
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Run inference on the current frame
    results = model(frame)

    # Visualize the detections
    annotated_frame = results[0].plot()

    # Display the frame
    cv2.imshow("YOLOv8 Real-Time Webcam Detection", annotated_frame)

    # Press 'q' to quit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
print("Webcam detection stopped.")
```
